```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.txt', sep='\t')
ames = pd.read_csv('ames_housing.txt', sep='\t')
titanic = pd.read_csv('titanic.txt', sep='\t')
batting = pd.read_csv('batting.csv', sep=',')
```

## Problem 1: Iris Dataset

```python
iris.head(8)
```

```python
iris.describe()
```

```python
iris_means = iris.iloc[:,:].groupby('species').mean()
iris_means
```

```python
iris_colors = ['lavender', 'darkorchid', 'violet']
iris_columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
for i in range(len(iris_columns)):
    plt.figure(figsize=[8,6])
    plt.subplot(2,2,i+1)
    plt.bar(x=iris_means.index, height=iris_means.iloc[:,i], color=iris_colors, edgecolor='k')
    plt.xlabel('Flower Species')
    plt.ylabel('Average ' + iris_columns[i])
    plt.title('Mean ' + iris_columns[i])
plt.tight_layout()
plt.plot()
```

## Problem 7: Batting Data – Batting Averages

```python
batting_avg = batting.loc[:,['teamID','yearID','H','AB']].groupby(['teamID','yearID']).sum()
batting_avg.loc[:,'BA'] = batting_avg.loc[:,'H'] / batting_avg.loc[:,'AB']
batting_avg.sort_values(by='BA', ascending=False)
batting_avg.head(10)
```

## Problem 8: Batting Data – Cardinals vs. Cubs

```python
mask = np.logical_and(batting.loc[:,'teamID'] == 'SLN',batting.loc[:,'yearID'] >= 1900 )
stl_batting = batting.loc[mask,['yearID','H','AB','HR']].groupby('yearID').sum()
stl_batting.loc[:,'BA'] = stl_batting.loc[:,'H'] / stl_batting.loc[:,'AB']

maskchi = np.logical_and(batting.loc[:,'teamID'] == 'CHN',batting.loc[:,'yearID'] >= 1900 )
chi_batting = batting.loc[maskchi,['yearID','H','AB','HR']].groupby('yearID').sum()
chi_batting.loc[:,'BA'] = chi_batting.loc[:,'H'] / chi_batting.loc[:,'AB']
```

```python
plt.figure(figsize=[12,4])
plt.subplot(1,2,1)
plt.plot(stl_batting.index, stl_batting.loc[:,'BA'], color='red', label='Cardinals')
plt.plot(chi_batting.index, chi_batting.loc[:,'BA'], color='b', label='Cubs')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Batting Average')
plt.title('Batting Average by Year')
plt.subplot(1,2,2)
plt.plot(stl_batting.index, stl_batting.loc[:,'HR'], color='red', label='Cardinals')
plt.plot(chi_batting.index, chi_batting.loc[:,'HR'], color='b', label='Cubs')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Home Runs')
plt.title('Home Runs by Year')
plt.show()
```
